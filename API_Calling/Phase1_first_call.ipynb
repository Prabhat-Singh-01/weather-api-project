```python
# phase1_first_call.py
# Goal: Make our first API call and understand the response

import requests   # for making HTTP requests (like a browser, but in Python)
import json       # for pretty-printing JSON data

# --- CONFIGURATION ---
API_KEY ='e365ed8fdcf3fbe95aba8de8c70d2bc5'
CITY = "Lucknow"                # let's start with a city close to home!
UNITS = "metric"                # metric = Celsius

# --- BUILD THE URL ---
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

params = {
    "q": CITY,
    "appid": API_KEY,
    "units": UNITS
}
# using 'params' dict is cleaner than manually writing ?q=...&appid=...

# --- MAKE THE API CALL ---
print(f"Fetching weather data for {CITY}...")

response = requests.get(BASE_URL, params=params)

# --- CHECK IF IT WORKED ---
print(f"Status Code: {response.status_code}")
# 200 = success, 401 = bad API key, 404 = city not found

# --- LOOK AT THE RAW RESPONSE ---
data = response.json()   # converts JSON text → Python dictionary

print("\n--- RAW JSON RESPONSE ---")
print(json.dumps(data, indent=2))   # indent=2 makes it readable
```

```python
# --- EXTRACT WHAT WE ACTUALLY NEED ---
print("\n--- EXTRACTED DATA ---")

city_name    = data["name"]
country      = data["sys"]["country"]
temperature  = data["main"]["temp"]
feels_like   = data["main"]["feels_like"]
humidity     = data["main"]["humidity"]
pressure     = data["main"]["pressure"]
wind_speed   = data["wind"]["speed"]
condition    = data["weather"][0]["main"]         # "weather" is a list!
description  = data["weather"][0]["description"]

print(f"City       : {city_name}, {country}")
print(f"Temperature: {temperature}°C (feels like {feels_like}°C)")
print(f"Humidity   : {humidity}%")
print(f"Pressure   : {pressure} hPa")
print(f"Wind Speed : {wind_speed} m/s")
print(f"Condition  : {condition} — {description}")
```

```python
# --- SAFE API CALL WITH ERROR HANDLING ---
response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    data = response.json()
    print("Success! Data received.")

elif response.status_code == 401:
    print("ERROR: Invalid API key. Check your key or wait 30 mins if new.")

elif response.status_code == 404:
    print(f"ERROR: City '{CITY}' not found. Check the spelling.")

else:
    print(f"ERROR: Something went wrong. Status code: {response.status_code}")
```
