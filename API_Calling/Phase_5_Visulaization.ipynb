```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('weather_analyzed.csv')
```

```python
df.head()
```

```python
# 1. Temperature by city

#PLOT

plt.figure(figsize=(14,6))
plt.plot(df['city'],
        df['temp_category'],
        marker = 'o',
        linewidth=2,
        color = 'green')
plt.title("Temperature Comparison Across 15 Cities")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")
plt.tight_layout()
plt.show
```

```python
df.head()
```

```python
# 2.  HUMIDITY vs TEMPERATURE

plt.figure(figsize=(10,7))
sns.scatterplot(data=df,
               x='city',
               y='temp_c',
               hue='humidity_level',
               palette='deep')
plt.xticks(rotation=45)
plt.title('Humidity vs Temperature')
plt.tight_layout()
plt.show
```

```python
df.head()
```

```python
# 3. WEATHER CONDITIONS

plt.figure(figsize=(14,6))

condition_counts = df['condition'].value_counts()
plt.pie(condition_counts.values,
       labels=condition_counts.index,
         autopct="%1.0f%%",
       )
plt.title("Average Temperature by Weather Condition")
plt.xlabel("Condition")
plt.ylabel("Avg Temperature (°C)")
plt.tight_layout()
plt.show()
```

```python
#4. Comfort index

plt.figure(figsize=(12,7))
sns.barplot(data=df,
           x='city',
           y='comfort_index',
           palette='Set3')
plt.title('City Comfort Index Ranking')
plt.tight_layout()
plt.show
```

```python
# 5. CORRELATION HEATMAP

plt.figure(figsize=(10,8))


numeric_cols = ["temp_c", "feels_like_c", "feels_like_diff",
                "humidity", "pressure_hpa", "wind_speed_mps", "comfort_index"]

corr = df[numeric_cols].corr().round(2)

sns.heatmap(corr,
            annot=True,           # show numbers inside boxes
            cmap="RdYlGn")        # red=negative, green=positive

plt.titl
```
